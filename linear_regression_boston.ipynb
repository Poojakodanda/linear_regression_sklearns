```python
## Independent and Dependent Features

x = dataset.iloc[0:,:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state =10)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_test
```

```python
## standardize or feature scaling the datasets

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
x_train = scaler.fit_transform(x_train) # 
```

```python
x_test = scaler.transform(x_test) # fit_transform is not used to avoid data leakage, mean and standard deviation is not calucated here, we are using mean and standard devivation of train data
```

```python
x_train
```

```python
x_test
```

####  Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression() # when we do this it takes all the default values into it
```

```python
regression        # LinearRegression() is a object
```

```python
# fit just to find out parameter 
# fit_transform to transorming data by applying formula on complete data and changing data 
```

```python
regression.fit(x_train, y_train)
```

```python
## print the coefficients and the intercept

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## Prediction for the test data

reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

#### Assumptions of Linear Regression

```python
plt.scatter(y_test, reg_pred) # we are checking relationship b/w true vale and predicted value. if we get linear relationship then our model is working good
plt.xlabel("Test Truth Data")
plt.ylabel("Test predicted Data")
```

```python
## residuals

residuals = y_test - reg_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind = 'kde') # if we get normally distribted curve, but we have right skewed we have check that. but other than that it is good
```

```python
## scatter plot with predictions and residual 
## uniform distribution

plt.scatter(reg_pred, residuals) # no shape no linearity , model is good
```

```python
## performance metrics

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

#### R square and adjusted R square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
## adjusted R square
## display adjusted r squared

1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
